```python
import os
import gc

import pandas as pd

from FoNN.similarity_search_dev import PatternSimilarity
```

```python
# find dirs for all tune families
root_dir = '/Users/dannydiamond/NUIG/Polifonia/thesession/test_corpus/tune_fams_ground_truth'
tune_family_names = [filename for filename in os.listdir(root_dir) if filename.endswith('.csv')]
print(tune_family_names)
```

```python
tune_families =  {family_name: None for family_name in tune_family_names}
for fam in tune_families:
    # Find and list all tunes in target tune family
    input_data = pd.read_csv(f'{root_dir}/{fam}', index_col=0)
    members = input_data['title'].tolist()
    tune_families[fam] = members

# reformat tune family names

trimmed_names = [n[:-4] for n in tune_family_names]
tune_families = dict(zip(trimmed_names, tune_families.values()))

for i, (k, v) in enumerate(tune_families.items()):
    print(f"\n\nTune family {i + 1}: {k}")
    for tune in v:
        print(tune)
    print(f"Total tunes: {len(v)}\n")
```

```python
# set up note-level instance of similarity experiment class
basepath = "//Users/dannydiamond/NUIG/Polifonia/thesession/feat_seq_corpus_apr23/note_pattern_corpus_3_12/"
patterns = basepath + 'patterns.npy'
titles = basepath + 'titles.npy'
tfidf = basepath + "tfidf_matrix.npz"
freq = basepath + "freq_matrix.npz"
tfidf_similarity = basepath + "tfidf_vector_cos_sim.mm"
feat_seq_path = "/Users/dannydiamond/NUIG/Polifonia/thesession/feat_seq_corpus_mar23/feat_seq_note"

test = PatternSimilarity(titles_path=titles, patterns_path=patterns, feat_seq_path=feat_seq_path, feature='diatonic_scale_degree')
# test.n = 6
# test.freq_matrix = freq
# test.tfidf_matrix = tfidf
# test.weighted_hamming_threshold = 1
test.tfidf_vector_cos_similarity_matrix = tfidf_similarity
```

```python
# Run TFIDF @ note-level: Done

for fam in tune_families:
    # Update this path for accent- or note-level input data.
    results_root_dir = f"/Users/dannydiamond/NUIG/Polifonia/thesession/similarity_results/large_scale_comparison_exp_4/note/{fam}"
    for tune in tune_families[fam]:
        test.query_tune = tune
        # raw pattern similarity results are stored here:
        test.out_dir = f"{results_root_dir}/{tune}"
        test.find_similar_tunes(mode='tfidf')
```
