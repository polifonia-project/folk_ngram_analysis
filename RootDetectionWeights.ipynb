# Root Detection - weights

Several well-known key detection algorithms work by comparing the histogram of pitch classes in the piece with a reference histogram or weighting of pitch classes, which is indicative of a particular key. This approach was proposed by Krumhansl and Schmuckler, and several different reference histograms have been proposed in the literature. Several are available in the `music21` software. In this short notebook we visualise these reference histograms.

```python
import matplotlib.pyplot as plt
```

```python
from music21.analysis.discrete import *
```

```python
pitches = "C C# D D# E F F# G G# A A# B".split()
```

```python
wtss =  (AardenEssen, BellmanBudge, KrumhanslKessler, KrumhanslSchmuckler, SimpleWeights, TemperleyKostkaPayne)
names = ("Aarden-Essen", "Bellman-Budge", "Krumhansl-Kessler", "Krumhansl-Schmuckler", "Sapp", "Temperley-Kostka-Payne")
for wts, name in zip(wtss, names):
    for maj_min in ("major", "minor"):
        w = wts().getWeights(maj_min)
        print(wts.__name__, w)
        plt.bar(x=pitches, height=w)
        plt.xlabel(f"{name} {maj_min}")
        plt.savefig(f"wts_{wts.__name__}_C_{maj_min}.pdf")
        plt.show()
        plt.close()
    
```
